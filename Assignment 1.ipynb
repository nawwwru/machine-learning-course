## 2
What is the cumulative distribution function (CDF) that corresponds to the following probability distribution function? Please state the value of the CDF for all possible values of $x$.

$f(x) = \begin{cases}
    \frac{1}{3} & 0 < x < 3 \\
    0           & \text{otherwise}
    \end{cases}$

**ANSWER**



As we know, CDF is the integral of PDF. Thus,  
when $x<0$,
$$F(x) = \int_{-\infty}^x 0 dx = 0$$  
when $0\leq x \leq3$,
$$F(x) = 0 + \int_{0}^x \frac{1}{3} dx = 0 + \frac{x}{3} = \frac{x}{3}$$  
when $x>3$,  
$$F(x) = 1$$

```python
import numpy as np
import time

# Generate the random samples
rand_arr = np.random.randn(10000000)

# Compute the sum of squares the non-vectorized way (using a for loop)
sum_1 = 0
rand_arr = rand_arr ** 2
start_1 = time.clock()
for i in rand_arr:
    sum_1 = i + sum_1
time_1 = (time.clock()-start_1)
# Compute the sum of squares the vectorized way (using numpy)
sum_2 = 0
start_2 = time.clock()
one_arr = np.ones(10000000)
sum_2 = np.dot(rand_arr,one_arr)
time_2 = (time.clock() - start_2)
# Print the results
print("Time [sec] (non-vectorized): {:f}".format(time_1))
print("Time [sec] (vectorized): {:f}".format(time_2))
print("The vectorized code is {:f} times faster than the non-vectorized code".format(time_1/time_2))
```

The function below computes the function $f(x,y) = x^2 - 2 y^2$ and determines whether this quantity is above or below a given threshold, `thresh=0`. This is done for $x,y \in \{-4,4\}$, with 2,000 samples in each direction.

(a) Vectorize this code and demonstrate (as in the last exercise) the speed increase through vectorization and (b) plot the resulting data - both the function $f(x,y)$ and the thresholded output - using [`imshow`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html?highlight=matplotlib%20pyplot%20imshow#matplotlib.pyplot.imshow) from `matplotlib`.

*Hint: look at the `numpy` [`meshgrid`](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.meshgrid.html) documentation*

```python
import numpy as np
import time
import matplotlib.pyplot as plt

# Initialize variables for this exerise
x = np.arange(-4,4,0.004)
x_sq = x**2
y = np.arange(-4,4,0.004)
y_sq_dw = 2*y**2
thres = 0
f1 = np.zeros(shape=(2000,2000))
f2 = np.zeros(shape=(2000,2000))
# Nonvectorized implementation
start1 = time.clock()
for i in range(1,2000):
    for j in range(1,2000):
        f1[i,j] = x_sq[i] - y_sq_dw[j]
time1 = (time.clock()-start1)
# Vectorized implementation
start2 = time.clock()
[X,Y] = np.meshgrid(x_sq,y_sq_dw)
f2 = X-Y
time2 = (time.clock()-start2)
f3 = f2.copy()
f3[f3 < 0] = 0
f3[f3 > 0] = 1
# Print the time for each and the speed increase
print("Time [sec] (non-vectorized): %f" % time1)
print("Time [sec] (vectorized): %f"  % time2)
print("The vectorized code is %f times faster than the non-vectorized code" % (time1/time2))
# Plot the result
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.set_title('All colormap')
ax1.imshow(f2)
ax2 = fig.add_subplot(122)
ax2.set_title('Threshold plot')
ax2.imshow(f3,cmap=plt.cm.gray)
plt.show()
```

*Table 1. Dataset containing n=5 observations*

| $x_i$ | 0.0 | 0.5 | 1.0 | 1.5 | 2.0 |
|-|-|-|-|-|-|
| $y_i$ | 6 | 7 | 8 | 4 | 1 |


**ANSWER**  
We can treat the function like:  
 $$f(x) = \begin{cases}
                6       & 0 \leq x < 0.5  \\
                7        & 0.5 \leq x < 1 \\
                8        & 1 \leq x <1.5 \\
                4       & 1.5 \leq x < 2 \\
                1       & 2 \leq x < 2.5
            \end{cases}$$
 Thus, the integral of $f(x)$ can be calculated as: \\
 $$\int_0^{2.5} f(x) dx = \int_0^{0.5} 6 dx + \int_{0.5}^1 7 dx + \int_1^{1.5} 8 dx + \int_{1.5}^2 4 dx + \int_2^{2.5} 1 dx = 13$$
